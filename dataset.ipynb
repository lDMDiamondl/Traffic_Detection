```python
import kaggle
import os

dataset_name = 'meowmeowmeowmeowmeow/gtsrb-german-traffic-sign'
download_path = './data'
```

```python
# 데이터가 없을 때만 다운로드 (중복 방지)

if not os.path.exists(download_path) or len(os.listdir(download_path)) == 0:
    os.makedirs(download_path, exist_ok=True)
    try:
        print(f"데이터셋 다운로드 중: {dataset_name}...")
        kaggle.api.dataset_download_files(dataset_name, path=download_path, unzip=True)
        print("다운로드 및 압축 해제 완료")
    except Exception as e:
        print(f"다운로드 실패: {e}")
else:
    print("이미 데이터가 존재하여 다운로드를 스킵합니다.")

target_root = None

print("\n--- 폴더 구조 출력 ---")
for root, dirs, files in os.walk(download_path):
    level = root.replace(download_path, '').count(os.sep)
    indent = ' ' * 4 * level
    folder_name = os.path.basename(root)
    if folder_name: # 루트 경로 출력용
        print(f"{indent}{folder_name}/")

    if 'train' in [d.lower() for d in dirs] and 'test' in [d.lower() for d in dirs]:
        target_root = root

print("-" * 25)
if target_root:
    print(f"검출된 데이터셋 루트: {target_root}")
else:
    print("데이터셋 루트를 찾지 못했습니다. 구조를 다시 확인해주세요.")
```
